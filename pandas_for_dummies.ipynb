```python
raw_75 = diamonds.sample(frac=0.75, random_state=99)
raw_75 #same as above but only 75%
```

```python
raw_25 = diamonds.loc[~diamonds.index.isin(raw_75.index), :] #locates on the diamonds df what's NOT raw_75 via indexing
raw_25
```

```python
training_df = raw_75
test_df = raw_25

#just wanted to mess around and show how this can then turn into the first steps of a typical lin.reg task
```

```python
x_dup = training_df.duplicated().sum()
y_dup = test_df.duplicated().sum()

print(f'Training duplicate = {x_dup}')
print(f'Test duplicate = {y_dup}')
```
